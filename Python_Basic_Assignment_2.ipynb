# Que. 11 In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
# range(10) :- for loop will take the values between 0 to 10. Initial value will be considered as '0' of not given.
lst1 = []
for i in range(10):
    lst1.append(i)
print("List for given range(10) :\n",lst1,'\n')


# range(0, 10) :- here, for loop will take initial point as 0 and end point will be 10. It will not return last value
# i.e 10
lst2 = []
for i in range(0,10):
    lst2.append(i)
print("List for given range(0,10) :\n",lst2,'\n')

# range(0,10,1) :- here we have assigned the step size = 1. Therefore, for loop will be iterated over the step size
# from initial value to the final. 
lst3 = []
for i in range(0,10,1):
    lst3.append(i)
print("List for given range(0,10,1) :\n",lst3)
```
