# 9. How do you make a function variable refer to the global variable?

```python
# 1. For referring global variable by function variable, we have to use 'global' keyword while expression
# for eg.

x = 'global x'

def local_var():
    global x
    x = 'local x'
    return x

local_var()                   # It will return the value of function variable x, not global x
```
